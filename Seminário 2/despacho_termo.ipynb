```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DESPACHO TERMO/GERACAO_TERMICA_DESPACHO_2021.csv', sep=';', parse_dates=['din_instante'], index_col='din_instante')
```

```python

curva_carga_series = pd.read_parquet('CURVA DE CARGA/curva de carga.parquet').loc[lambda x: x.index.year==2021].sum(axis=1)
```

```python
curva_carga_series.resample('W').mean().plot()
```
