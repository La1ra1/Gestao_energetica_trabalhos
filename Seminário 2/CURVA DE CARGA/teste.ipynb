```python
from glob import glob
import locale
import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CURVA_CARGA_2023.csv', sep=';', index_col='din_instante', parse_dates=['din_instante'])
```

```python
aux = pd.DataFrame(columns=['N', 'NE', 'S', 'SE'])

for reg in ['N', 'NE', 'S', 'SE']:
    aux[reg] = df.query("id_subsistema == @reg")['val_cargaenergiahomwmed'].dropna()
```

```python
principal = pd.read_parquet('curva de carga.parquet')
```

```python
curva_carga_atual = pd.concat([principal, aux])[lambda x: ~x.index.duplicated(keep='first')]
curva_carga_atual
```
